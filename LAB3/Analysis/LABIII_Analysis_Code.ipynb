```python
# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numr = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denr = sum([xi**2 for xi in X]) - n * xbar**2

    b = numr / denr
    a = ybar - b * xbar

    #print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b
```

```python
# Error calculation for data collected in ISEC building
east_ISEC_1 = easting_closed_move[:300]
north_ISEC_1 = northing_closed_move[:300]
a1, b1 = best_fit(east_ISEC_1, north_ISEC_1)

plt.scatter(east_ISEC_1, north_ISEC_1, color = "Blue")
yfit = [a1 + b1 * xi for xi in east_ISEC_1]
plt.plot(east_ISEC_1, yfit, color = "Black")



model = LinearRegression().fit(east_ISEC_1.reshape(-1,1),north_ISEC_1.reshape(-1,1))
print("Mean Absolute Error (Path 1) = ", mean_absolute_error(north_ISEC_1.reshape(-1,1), model.predict(east_ISEC_1.reshape(-1,1))))


# Error calculation for data collected in ISEC building
east_ISEC_2 = easting_closed_move[300:400]
north_ISEC_2 = northing_closed_move[300:400]
a2, b2 = best_fit(east_ISEC_2, north_ISEC_2)

plt.scatter(east_ISEC_2, north_ISEC_2, color = "Blue")
yfit = [a2 + b2 * xi for xi in east_ISEC_2]
plt.plot(east_ISEC_2, yfit, color = "Black")



model = LinearRegression().fit(east_ISEC_2.reshape(-1,1),north_ISEC_2.reshape(-1,1))
print("Mean Absolute Error (Path 2) = ", mean_absolute_error(north_ISEC_2.reshape(-1,1), model.predict(east_ISEC_2.reshape(-1,1))))

# Error calculation for data collected in ISEC building
east_ISEC_3 = easting_closed_move[400:700]
north_ISEC_3 = northing_closed_move[400:700]
a3, b3 = best_fit(east_ISEC_3, north_ISEC_3)

plt.scatter(east_ISEC_3, north_ISEC_3, color = "Blue")
yfit = [a3 + b3 * xi for xi in east_ISEC_3]
plt.plot(east_ISEC_3, yfit, color = "Black")



model = LinearRegression().fit(east_ISEC_3.reshape(-1,1),north_ISEC_3.reshape(-1,1))
print("Mean Absolute Error (Path 3) = ", mean_absolute_error(north_ISEC_3.reshape(-1,1), model.predict(east_ISEC_3.reshape(-1,1))))


# Error calculation for data collected in ISEC building
east_ISEC_4 = easting_closed_move[700:]
north_ISEC_4 = northing_closed_move[700:]
a4, b4 = best_fit(east_ISEC_4, north_ISEC_4)

plt.scatter(east_ISEC_4, north_ISEC_4, color = "Blue")
yfit = [a4 + b4 * xi for xi in east_ISEC_4]
plt.xlabel('UTM_EASTING')
plt.ylabel('UTM_NORTHING')
plt.title('NORTHING vs EASTING (ISEC)')
plt.plot(east_ISEC_4, yfit, color = "Black")



model = LinearRegression().fit(east_ISEC_4.reshape(-1,1),north_ISEC_4.reshape(-1,1))
print("Mean Absolute Error (Path 4) = ", mean_absolute_error(north_ISEC_4.reshape(-1,1), model.predict(east_ISEC_4.reshape(-1,1))))

```
